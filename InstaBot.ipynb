```python
import re
hashTags = {}
words = {}
driver = webdriver.Chrome()
driver.get("https://www.instagram.com/")
driver.maximize_window()
driver.implicitly_wait(5)
username = driver.find_element(By.XPATH, "//input[@aria-label = 'Phone number, username, or email']")
username.send_keys("anubhav_75")
password = driver.find_element(By.XPATH, "//input[@aria-label = 'Password']")
password.send_keys("Hazzard$11008")
login = driver.find_element(By.XPATH, "//div[text() = 'Log in']")
login.click()
for i in handles_ :
    search = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, "//*[local-name() = 'svg' and @aria-label = 'Search & Explore']"))
    )
    search.click()
    search_input = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, "//input[@aria-label = 'Search input']"))
    )
    search_input.send_keys(i)
    time.sleep(5)
    result = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, "//div[@class = 'x9f619 x1n2onr6 x1ja2u2z x1qjc9v5 x78zum5 xdt5ytf x1iyjqo2 xl56j7k xeuugli']"))
    )
    result.click()
    driver.execute_script("window.scrollTo(0, 0);")
    time.sleep(2)
    posts = driver.find_elements(By.XPATH, "//div[@class = 'x1lliihq x1n2onr6 xh8yej3 x4gyw5p xfllauq xo2y696 x11i5rnm x2pgyrj']")
    l = 0
    for j in posts:
        j.click()
        try:
            caption = driver.find_element(By.XPATH, "//h1[@class = '_ap3a _aaco _aacu _aacx _aad7 _aade']").text
            caption = caption.replace("\n", " ")
            print(caption)
            emoji_pattern = re.compile(
            "["
            "\U0001F600-\U0001F64F"  # Emoticons
            "\U0001F300-\U0001F5FF"  # Symbols & Pictographs
            "\U0001F680-\U0001F6FF"  # Transport & Map Symbols
            "\U0001F700-\U0001F77F"  # Alchemical Symbols
            "\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
            "\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
            "\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
            "\U0001FA00-\U0001FA6F"  # Chess Symbols
            "\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
            "\U00002702-\U000027B0"  # Dingbats
            "\U000024C2-\U0001F251" 
            "]+", 
            flags=re.UNICODE
            )
            caption = emoji_pattern.sub(r'', caption)
            caption = caption.split(" ")
        except:
            driver.find_element(By.XPATH, "//*[local-name() = 'svg' and @aria-label = 'Close']").click()
            l = l + 1
            continue
        for k in caption:
            if k.startswith("#"):
                if k in hashTags :
                    hashTags[k] = hashTags[k] + 1
                else:
                    hashTags[k] = 1
            else:
                if k in words :
                    words[k] = words[k] + 1
                else:
                    words[k] = 1
        driver.find_element(By.XPATH, "//*[local-name() = 'svg' and @aria-label = 'Close']").click()
        l = l + 1
        if l == 10:
            break
```
